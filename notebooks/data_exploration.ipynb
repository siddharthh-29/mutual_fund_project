```python
import pandas as pd
import glob
import os
from IPython.display import display

# target your raw data folder
raw_data_dir = r"C:\Users\sivap\mutual_fund_project\data\raw"
csv_files = glob.glob(f"{raw_data_dir}/*.csv")

print(f"Found {len(csv_files)} datasets to inspect.\n")

for file in csv_files:
    filename = os.path.basename(file)
    print(f"{'='*40}")
    print(f"DATASET: {filename}")
    print(f"{'='*40}")
    
    try:
        df = pd.read_csv(file)
        
        # 1. basic shape and memory
        print(f"Rows: {df.shape[0]} | Columns: {df.shape[1]}")
        
        # 2. check for anomalies (missing data & duplicates)
        missing = df.isnull().sum()
        missing_cols = missing[missing > 0]
        duplicates = df.duplicated().sum()
        
        if not missing_cols.empty:
            print("\nWARNING - Missing Values Found:")
            print(missing_cols.to_string())
        else:
            print("\nClean: No missing values.")
            
        if duplicates > 0:
            print(f"WARNING - Duplicates: {duplicates} duplicate rows found.")
            
        # 3. data types (critical for tomorrow's processing)
        print("\nData Types:")
        print(df.dtypes.to_string())
        
        # 4. preview the actual bits of data
        print("\nData Preview:")
        display(df.head(3))
        print("\n\n")
        
    except Exception as e:
        print(f"Failed to read {filename}: {e}\n")
```
